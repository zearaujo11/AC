# Teams Post Dataset Metrics

## Section: Import Data
This section focuses on import libraries and players_teams dataset.
The process involves:
- import pandas and matplotlib.pyplot libraries
- import teams_post dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

teams_post = pd.read_csv('../basketballPlayoffs/teams_post.csv', delimiter=",")
```

## Section: Explore Dataset
This section focuses on find null values and extract informations about the dataset. The process involves:
- Find null values
- Calcule Win-loss ratios for teams on the post season

```python
teams_post.isnull().sum()
```

```python
# Calculate win-loss ratios
teams_post['Win-Loss Ratio'] = teams_post['W'] / (teams_post['W'] + teams_post['L'])

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
plt.barh(teams_post['tmID'], teams_post['Win-Loss Ratio'], color='skyblue')

# Add labels and title
plt.xlabel('Win-Loss Ratio')
plt.ylabel('Team ID (tmID)')
plt.title('Win-Loss Ratios for Teams on  Post-Season (based on tmID)')

# Show the chart
plt.show()
```
